## Описание проекта ##

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы
(например, Xboxили PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Легенда:**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.


- `Данные за 2016 год могут быть неполными.`

Отсутствует название, жанр игры, оценки пользователей, критиков и рейтинг игры. Удалим эти строки из-за минимального наличия информации.

#### Дата выхода игры ####

Невозможно предположить дату выхода игры. Необходимо запросить информацию из дополнительных источников или удалить эти строки (269 шт.), что составляет менее 2% и не повлияет на объем информации для анализа.

```python
df = df.dropna(subset = ['release_year']) #удаляем строки, где отсутствует дата релиза игры
```

```python
print(f'Количество строк: {df.shape[0]}, количество столбцов: {df.shape[1]}') #оценим объем датафрейма
```

#### Рейтинги ####

Оценим, зависит ли отсутствие рейтинга от даты выхода игры. Возможно, когда игры выходили рейтинги еще не были распространены.

```python
#оценим количество игр по имени без рейтинга от критиков сгруппировав данные по годам

ax1 = df[df['critic_score'].isna()].groupby('release_year').agg({'name':'count'})

#оценим количество игр по имени без рейтинга от пользователей сгруппировав данные по годам

ax2 = df[df['user_score'].isna()].groupby('release_year').agg({'name':'count'})

#оценим количество игр по имени сгруппировав данные по годам без дополнительных фильтров

ax0 = df.groupby('release_year').agg({'name':'count'})
```

```python
plt.figure(figsize=(15,5))
plt.plot(ax1, 'o-', ax2, 'x-', ax0, '-')
plt.title('Games count vs. Critics / Users rating count', fontsize=15)
plt.xlabel('Year', fontsize=12, color='blue')
plt.ylabel('Count', fontsize=12, color='blue')
plt.grid(True)
plt.legend(["critics","users","total"])
plt.show()
```

**Вывод:**
- На основе графиков можно сделать вывод, что до 1998 года рейтинги критиков и пользователей минимальны по количеству, вероятно, по причине развития этого направления активности значительно позднее.
- Начиная с 2000 года можно отметить бум и резкий рост количества выпущенных игр, а с ними и рост количества отзывов экспертов на эти релизы c 2001. С 2009 года идет спад в игровой индустрии и стабилизация начиная с 2012 года.
- В тоже время доля игр без отзывов от критиков и/или пользователей остается очень высокой, чтобы просто удалить эти строки датафрейма или заполнить медианами / средними значениями, тем самым "засорив" исходные данные необоснованными допущениями о возможном рейтинге той или иной игры.

Заполним все пропущенные значения значением "-1", чтобы:
   - отслеживать эти строки в дальнейшем;
   - поменять тип данный на целочисленный на следующем этапе.

```python
#заполнение всех пропусков по рейтингам от критиков на "-1"

df['critic_score'] = df['critic_score'].fillna(-1)

#заполнение всех пропусков по рейтингам от пользователй на "-1" с изменением 'tbd' на (-1)

df.loc[df['user_score'] == "tbd",'user_score'] = -1
df['user_score'] = df['user_score'].fillna(-1)

#заполнение всех пропусков по рейтингу от организации ESRB на "unknown"

df['rating'] = df['rating'].fillna("unknown")
```

```python
df.head()
```

```python
print(f'Количество строк: {df.shape[0]}, количество столбцов: {df.shape[1]}') #оценим объем датафрейма
```

### Изменение типа данных ###

Использование типа данных 'int64' позволяет экономить ресурсы (память) в сравнении с типом данных 'float64'.

```python
#замена типа на целое число 

df['release_year'] = df['release_year'].astype('int')
df['user_score'] = df['user_score'].astype('float64')
```

## Проверка гипотез ##

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые ###

`Гипотеза о равенстве средних двух генеральных совокупностей:`
- Гипотеза H0: Средние пользовательские рейтинги платформ Xbox One и PC равны (H0 всегда со знаком "=").
- Гипотеза H1: Средние пользовательские рейтинги платформ Xbox One и PC НЕ равны.

*Рассмотрим три аспекта, которые надо проверить, чтобы понять, можно ли применять t-тест:*

- [x] Генеральные совокупности не должны зависеть друг от друга.
    - Да, в нашем случае пользователи разных платформ независимы друг от друга.
- [x] Выборочные средние должны быть нормально распределены.
    - А если выборка всего одна, то её среднее должно иметь нормальное распределение для разных выборок этого размера из определённой генеральной совокупности.
    - Благодаря ЦПТ, если размер выборки составляет хотя бы несколько десятков значений, выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены нормально вокруг истинного среднего этой совокупности. Утверждение верно, даже если сама генеральная совокупность не распределена нормально.

```python
#соберем два датафрейма по пользователям двух разных платформ
df_actual_users_xone = df_actual.query('platform == "XOne"')['user_score']
df_actual_users_pc = df_actual.query('platform == "PC"')['user_score']
```

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results_users_platform = st.ttest_ind(
    df_actual_users_xone, 
    df_actual_users_pc)

print('Средний рейтинг пользователей XOne"', round(df_actual_users_xone.mean(),2))
print('Средний рейтинг пользователей PC', round(df_actual_users_pc.mean(),2))
print('p-значение: ', results_users_platform.pvalue)

if results_users_platform.pvalue < alpha:
    print()
    print("Отвергаем нулевую гипотезу")
else:
    print()
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод:** Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно >20% - большая вероятность (выше уровня статистической значимости), чтобы делать вывод о значимом различии между оценками на платформе XOne и PC.

### Средние пользовательские рейтинги жанров Action и Sports разные ###

`Гипотеза о равенстве средних двух генеральных совокупностей:`
- Гипотеза H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые / равны (H0 всегда со знаком "=").
- Гипотеза H1: Средние пользовательские рейтинги жанров Action и Sports НЕ равны.

```python
#соберем два датафрейма по пользователям двух разных жанров
df_actual_users_action = df_actual.query('genre == "Action"')['user_score']
df_actual_users_sports = df_actual.query('genre == "Sports"')['user_score']
```

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results_users_genre = st.ttest_ind(
    df_actual_users_action, 
    df_actual_users_sports)

print('Средний рейтинг по жанру "Action"', round(df_actual_users_action.mean(),2))
print('Средний рейтинг по жанру "Sports"', round(df_actual_users_sports.mean(),2))
print('p-значение: ', results_users_genre.pvalue)

if results_users_genre.pvalue < alpha:
    print()
    print("Отвергаем нулевую гипотезу")
else:
    print()
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод:** Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно очень маленькая - меньше уровня статистической значимости, чтобы делать вывод о том, что рейтинги одинаковы. Принимаем гипотезу H1 -  cредние пользовательские рейтинги жанров Action и Sports НЕ равны / разные.
